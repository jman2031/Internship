# Car Price Scrapping

Price of used cars have been scrapped from the website https://www.cars24.com/. All the metro cities have been included in this process. The files have been both saved in .csv and.xlsx formats. 

```python
# Importing Libraries
import pandas as pd
from bs4 import BeautifulSoup
import requests
import selenium
from selenium import webdriver
import time
import warnings                                            
warnings.filterwarnings('ignore')

from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import NoSuchElementException,ElementNotInteractableException,WebDriverException
from selenium.common.exceptions import SessionNotCreatedException
```

```python
import nltk
nltk.download('punkt')
import nltk
from nltk.tokenize import word_tokenize
from nltk.tokenize import regexp_tokenize
```

```python
url = "https://www.cars24.com/"
```

```python
# Getting respnse from the given page link
page = requests.get(url)
print(page)
soup = BeautifulSoup(page.content)
```

```python
driver = webdriver.Chrome(r"C:\Users\jyuth\Chrome driver 103\chromedriver.exe")
driver.get(url)
driver.maximize_window()
```

```python
viewall=driver.find_element_by_xpath('/html/body/div[1]/div[1]/header/div/div[2]/div[2]/button')
viewall.click()
```

```python
#scrolling cursor down in order to get number of details added on the page
for _ in range(600):
    driver.execute_script("window.scrollBy(0,600)")
```

```python
Model=[]
#scrapping Model
model=driver.find_elements_by_xpath("//h2[@class='_3FpCg']")
for i in model:
    Model.append(i.text)
Model

#scrapping Transmission
Transmission=[]
T=driver.find_elements_by_xpath("//p[@class='cvakB']/span")
for i in T:
    Transmission.append(i.text)
Transmission

#scrapping Distance
Kms=[]
k=driver.find_elements_by_xpath("//ul[@class='bVR0c']/li[1]")
for i in k:
    Kms.append(i.text)
Kms

#scrapping ownership
owner=[]
o=driver.find_elements_by_xpath("//ul[@class='bVR0c']/li[2]")
for i in o:
    owner.append(i.text)
owner

#scrapping fuel type
fuel=[]
f=driver.find_elements_by_xpath("//ul[@class='bVR0c']/li[3]")
for i in f:
    fuel.append(i.text)
fuel

#csrapping Price
Price=[]
p=driver.find_elements_by_xpath("//div[@class='_7udZZ']/span")
for i in p:
    Price.append(i.text)
Price

#checking length of each scapped element
len(Model), len(Price), len(Transmission),len(fuel),len(Kms),len(owner)
```

```python
#Making Data frame
df1=pd.DataFrame({'Model':Model,'Transmission':Transmission,'Fuel':fuel,'Kms':Kms,'Owner':owner, 'Price':Price})
df1

#saving files to .csv and .xlsx formats
Cp1=df1.to_csv('Cp1.csv')
Cp1=df1.to_excel('Cp1.xlsx')
```
