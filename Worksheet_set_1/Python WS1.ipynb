#### 11. Write a python program to find the factorial of a number. 

```python
import math 
def factorial(f):
    return(math.factorial(f))

x=int(input('Enter a number:'))
      
print('Factorial of number', x, 'is', factorial(x))
```

#### 12. Write a python program to find whether a number is prime

```python
n=int(input("Enter number:"))

if n>1:
    for i in range(2,n):
        if (n%2)==0:        
            print(n, "Number is composite number")
            break
    else:
        print(n,"is prime number")
```

#### 13. Write a python program to check whether a given string is palindrome or not

```python
p=str(input("Enter name:"))

if (p==p[::-1]):
    print('It is a palindrome')
else:
    print('It is a word and not a palindrome')
```

#### 15. Write a python program to print the frequency of each of the characters present in a given string 

```python
from collections import Counter
  
# initializing string 
x = "I didnt know the soultion to this question, then I refered to the suggested url"
  
# using collections.Counter() to get count of each element in string 
y = Counter(x)
  
# printing result 
print ("Count of all characters in the sentence is :" + str (y))
```
