# Q2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
#get page request
page=requests.get('https://www.imdb.com/chart/top/')
page

soup=BeautifulSoup(page.content)
soup

#Find names,movies, ratings,year
Names=soup.find_all('td',class_="titleColumn")
print(Names)

movies=[]
for i in Names:
    for x in i.find_all('a'):
        movies.append(x.text)    
print(movies)

Rating=[]
for i in soup.find_all('td',class_="ratingColumn imdbRating"):
      Rating.append(i.text.replace("\n",""))
print(Rating)

Year=[]
for i in soup.find_all('span',class_="secondaryInfo"):
    Year.append(i.text)
print(Year)

#Check length
print(len(movies),len(Rating),len(Year))

#make data Frame
df=pd.DataFrame({'Movie Names':movies,'Rating':Rating,'Year of Release':Year})
df
```

```python
#dropping uncessary data
df1=df.drop(df.index[100:250])
df1

#save file in CSV
imdb=df1.to_csv('IMDB_top_100_movies')
imdb
```

# Q4 Write a python program to scrape product name, price and discounts from https://meesho.com/bags-ladies/pl/p7vbp .

```python
#get page request
page=requests.get('https://meesho.com/bags-ladies/pl/p7vbp')
page

soup=BeautifulSoup(page.content)
soup

#scrape name
Product_name=[]
for i in soup.find_all('p',class_="Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS"):
      Product_name.append(i.text)
print(Product_name)

#Scrape price
Price=[]
for i in soup.find_all('h5',class_="Text__StyledText-sc-oo0kvp-0 hiHdyy"):
      Price.append(i.text)
print(Price)

#Scrape discount
Discount=[]

for i in soup.find_all('span',class_="Text__StyledText-sc-oo0kvp-0 lnonyH"):
    Discount.append(i.text.split()[0])
print(Discount)

#check length
print(len(Product_name),len(Price),len(Discount))

#make data Frame
M=pd.DataFrame({'Product Name':Product_name,'Price':Price,'Discount on product':Discount})
M

#save file in CSV
Meesho=M.to_csv('Meesho_ladies_bag')
Meesho
```

# Q9) Write a python program to scrape mentioned details from dineout.co.in :
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
#Import the needed libraries

#get page request
page=requests.get('https://www.dineout.co.in/delhi-restaurants?city_name=delhi&limit=21&start=0&cityId=0&listing=1&tag%5B%5D=Dineout%20Pay&showAvailableTicket=0')

print(page)

#get page content
soup=BeautifulSoup(page.content)
soup

#for scraping restaurent name
RN =[]
for i in soup.find_all('a',class_="restnt-name ellipsis"):
    RN.append(i.text)
print(RN)

#for scraping cuisine
C=[]
for i in soup.find_all('span',class_="double-line-ellipsis"):
    C.append(i.text.split('|')[1])
print(C)

#for scraping Location
L=[]
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    L.append(i.text)
print(L)

#for scraping Ratings
r=[]
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    r.append(i.text)
print(r)

#for scraping images
Images=[]
for i in soup.find_all('img',class_="no-img"):
    Images.append(i['data-src'])
print(Images)

#find length
print(len(RN),len(C),len(L),len(r),len(Images))
```

```python
#Import the needed libraries

#get page request
page=requests.get('https://www.bewakoof.com/women-clothing/category-t-shirt')

print(page)

#get page content
soup=BeautifulSoup(page.content)
soup

#for scraping the product name
N =[]
for i in soup.find_all('h3'):
    N.append(i.text)
name=N[:10]
print(name)

#for scraping price
P=[]
for i in soup.find_all('span',class_="discountedPriceText"):
    P.append(i.text)
print(P)

#for scraping images
Images=[]
for i in soup.find_all('img',class_="productImgTag"):
    Images.append(i['src'])
print(Images)

print(len(name),len(P),len(Images))
```

```python
#Make dataframe and csv file

df=pd.DataFrame({'Product Name':name,'Price':P,'Image url':Images})
df

#csv file
B=df.to_csv("Bewakoof_women_tshirts")
B
```
