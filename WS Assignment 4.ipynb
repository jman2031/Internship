```python
#creating Empty lists:
DN=[]   #DataSetName
DT=[]   #DataType
Task=[]
AT=[]   #AttributeType
NoI=[]  #numberOfInstances
NoA=[]  #No of attribute
Year=[]

#For Scrapping DataSetName:
dn_tags = driver.find_elements_by_xpath('//td//p[@class="normal"]')
for i in dn_tags:
    DN.append(i.text)
Dn=DN[8::7]
DN=Dn[:622]

#For Scrapping DataSettype:
dt_tags = driver.find_elements_by_xpath('//td//p[@class="normal"]')
for i in dt_tags:
    DT.append(i.text)
Dt=DT[9::7]
DT=Dt[:622]

#For Scrapping Task:
t_tags = driver.find_elements_by_xpath('//td//p[@class="normal"]')
for i in t_tags:
    Task.append(i.text)
task=Task[10::7]
Task=task[:622]

#For Scrapping Attribute type:
at_tags = driver.find_elements_by_xpath('//td//p[@class="normal"]')
for i in at_tags:
    AT.append(i.text)
At=AT[11::7]
AT=At[:622]

#For Scrapping Number of Instances:
Noi_tags = driver.find_elements_by_xpath('//td//p[@class="normal"]')
for i in Noi_tags:
    NoI.append(i.text)
noi=NoI[12::7]
NOI=noi[:622]

#For Scrapping Number of Attributes:
Noa_tags = driver.find_elements_by_xpath('//td//p[@class="normal"]')
for i in Noa_tags:
    NoA.append(i.text)
noa=NoA[13::7]
NOA=noa[:622]

#For Scrapping Year:
y_tags = driver.find_elements_by_xpath('//td//p[@class="normal"]')
for i in y_tags:
    Year.append(i.text)
year=Year[14::7]
Year=year[:622]

print(len(Dn), len(Dt),len(task),len(At),len(noi),len(noa),len(year))

df10=pd.DataFrame({'DataSet Name':DN,'Data Type':DT,'Task':Task,'Attribute Type':AT,'Number of Instances':NOI,
                   'Number of Attributes':NOA,'Year':Year})
df10

#save to .csv
j=df10.to_csv('UCIData')
j
```
